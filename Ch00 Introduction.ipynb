# Short intro to pandas

A DataFrame is one of the major data types provided by pandas. It is a class thtat represents a spreadsheet, with cell values, column names and row index labels. 


```python
# create a dataframe from a list of dictionaries
import pandas as pd

data = [
    {'name': 'Lotte', 'age': 25, 'city': 'Antwerpen'},
    {'name': 'Bram', 'age': 30, 'city': 'Gent'},
    {'name': 'Fien', 'age': 35, 'city': 'Brugge'},
    {'name': 'Jasper', 'age': 40, 'city': 'Leuven'},
    {'name': 'Anke', 'age': 45, 'city': 'Mechelen'},
    {'name': 'Wout', 'age': 50, 'city': 'Hasselt'}
]

df = pd.DataFrame(data)
print(df)
```

```python
# save a dataframe to a csv file
df.to_csv('people.csv', index=False)
```

```python
# read a dataframe from a csv file
people = pd.read_csv('people.csv')
print(people)
```

```python
# elementary data analytics: what is the average age of the people in the dataframe?
average_age = people['age'].mean()
print(f"The average age of the people in the dataframe is: {average_age}")
```
