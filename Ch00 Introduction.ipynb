# Short intro to pandas

A DataFrame is one of the major data types provided by pandas. It is a class thtat represents a spreadsheet, with cell values, column names and row index labels. 


```python
# create a dataframe from a list of dictionaries
import pandas as pd

data = [
    {'name': 'Lotte', 'age': 25, 'city': 'Antwerpen'},
    {'name': 'Bram', 'age': 30, 'city': 'Gent'},
    {'name': 'Saskia', 'age': 40, 'city': 'Gent'},
    {'name': 'Fien', 'age': 35, 'city': 'Brugge'},
    {'name': 'Jasper', 'age': 40, 'city': 'Leuven'},
    {'name': 'Anke', 'age': 45, 'city': 'Mechelen'},
    {'name': 'Bart', 'age': 60, 'city': 'Leuven'},
    {'name': 'Wout', 'age': 50, 'city': 'Hasselt'}
]

df = pd.DataFrame(data)
print(df)
```

```python
tekst  = 'xxx'
print(tekst)
```

```python
# save a dataframe to a csv file
df.to_csv('people.csv', index=False, sep=';')
```

```python
# read a dataframe from a csv file
people = pd.read_csv('people.csv', sep=';')
print(people)
```

```python
# elementary data analytics: what is the average age of the people in the dataframe?
average_age = people['age'].mean()
print(f"The average age of the people in the dataframe is: {average_age}")
```

```python
# average age per city
average_age_per_city = people.groupby('city')['age'].mean()
print("The average age per city is:")
print(average_age_per_city)
```

Maak een functie genaamd `maak_spionnennaam(voornaam, lievelingskleur)` die een “spionnennaam” genereert.

Regels:

- Neem de eerste 3 letters van de voornaam.
- Neem de laatste 2 letters van de lievelingskleur.
- Voeg daartussen een willekeurig getal tussen 1 en 99 aan toe (altijd 2 cijfers, dus 01, 02, ..., 99).
- Zet alles achter elkaar in hoofdletters.

Voorbeeld:

- Invoer: voornaam = "Julia", lievelingskleur = "groen"
- Mogelijke uitvoer: "JUL47EN"
- Tip: Gebruik slicing, upper(), zfill() en random.randint().

Bonus:
- Vraag voornaam en lievelingskleur op aan de gebruiker
- Controleer dat de ingegeven strings geen cijfers bevatten. Indien wel, vraag dan opnieuw op tot ok. 

```python
import random

def bevat_cijfers(tekst):
    return any(teken.isdigit() for teken in tekst)

def maak_spionnennaam(voornaam, lievelingskleur):
    spion = voornaam[:3].upper()
    getal = random.randint(1, 99)  # geeft een getal tussen 1 en 99 (bijv. 47)
    spion = spion + str(getal).zfill(2) + lievelingskleur[-2:].upper()
    return f"{spion}"

def main():
    vnaam = "0"
    while bevat_cijfers(vnaam):
        vnaam = input('Geef voornaam (zonder cijfers):')

    lievelingskleur = "0"
    while bevat_cijfers(lievelingskleur):
        lievelingskleur = input('Geef lievelingskleur (zonder cijfers)')

    spionnennaam1 = maak_spionnennaam(vnaam, lievelingskleur)
    spionnennaam2 = maak_spoinnennaam('Johan','kleur')
    print(f"{spionnennaam}")

if __name__ == "__main__":
    main()
```
